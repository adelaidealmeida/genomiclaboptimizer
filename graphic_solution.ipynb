```python
import numpy as np
import matplotlib.pyplot as plt

# Repetio do grfico com rea azul e funo objetivo
x2_vals = np.linspace(0, 50, 500)
x3_tm = (100 - 2*x2_vals)/6    # Tempo de mquina
x3_rg = (200 - 5*x2_vals)/12   # Reagentes
x3_ht = (120 - 2*x2_vals)/5    # Horas tcnico
x3_obj = (166 - 4*x2_vals) / 10  # Z = 166 = 4x2 + 10x3 -> x3 = (166 - 4x2)/10

# Soluo tima
x2_opt, x3_opt = 4, 15

# Preencher regio vivel em azul
x2_grid, x3_grid = np.meshgrid(x2_vals, np.linspace(0, 30, 500))
feasible_region = (2*x2_grid + 6*x3_grid <= 100) & \
                  (5*x2_grid + 12*x3_grid <= 200) & \
                  (2*x2_grid + 5*x3_grid <= 120)

plt.figure(figsize=(10, 6))
plt.contourf(x2_vals, np.linspace(0, 30, 500), feasible_region, levels=1, colors=['#b0c4de'], alpha=0.8)

# Restries
plt.plot(x2_vals, x3_tm, label='Machine Time', color='orange')
plt.plot(x2_vals, x3_rg, label='Reagents (bottleneck)', color='green')
plt.plot(x2_vals, x3_ht, label='Technician hours', color='blue')

# Funo objetivo Z = 166
plt.plot(x2_vals, x3_obj, 'k--', label=r'Objective Function $4x_2 + 10x_3 = 166$')

# Soluo tima
plt.plot(x2_opt, x3_opt, 'ro', label='Optimal Solution (4, 15)')
plt.text(x2_opt + 0.5, x3_opt, ' (4, 15)', color='green')

# Configuraes
plt.xlabel('x2 (PCR)')
plt.ylabel('x3 (ChIP-seq)')
plt.title('Graphical Solution with Objective Function - Plane x1 = 0')
plt.xlim((0, 30))
plt.ylim((0, 30))
plt.grid(True)
plt.legend()
plt.show()
```
